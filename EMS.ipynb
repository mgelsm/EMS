```python
from IPython.core.display import HTML
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from bokeh.plotting import figure, ColumnDataSource
from bokeh.io import output_notebook, show
from bokeh.models import HoverTool

output_notebook()
```

```python
def display_interative_plot(df, x, y):
    if x == "MSsurC":
        tooltips=[
                ("Nom de l´établissement", "@index"),
                ("Lits", "@Lits"),
                ("Estimation m2", "@Esti_m2"),
                (x, "@"+x+" %"),
                (y, "@"+y+" k.-ChF")
            ]
    else:
        tooltips=[
                ("Nom de l´établissement", "@index"),
                ("Lits", "@Lits"),
                ("Estimation m2", "@Esti_m2"),
                (y, "@"+y+" k.-ChF")
            ]
    p = figure(title="EMS en fonction de "+ x + " et " + y)
    source = ColumnDataSource(df)

    hover = HoverTool(
            
        )

    p.circle(y=y, 
             x=x, source=source, size=10, alpha=0.5,
             color="blue", legend="EMS")

    p.add_tools(hover)
    p.legend.location = "top_left"
    show(p)
```

```python
df = pd.read_excel('EMS.xlsx')
df.columns = df.iloc[0,:]
df.drop([0,1], inplace=True)
df.set_index('Nom de l´établissement', inplace=True)
df.index.name='index'
df.head(3)
```

### /!\ Pour plus de simplicité dans le code, la colonne "Résultat d'exploitation" est renommée Benefice

```python
df.dropna(inplace=True)
df.rename(columns={"Résultat d´exploi-tation, sans déficit couvert (en milliers de francs)":'Benefice'}, inplace=True)
```

```python
a = df.iloc[:,3:]
a.replace({'-': 0, '*': 0}, inplace=True)
a = a.astype(float)
```

```python
# Je ne prends pas en compte les soins aigus car les patients ne se lèvent certainement pas 
df['Lits']=df['Nombre de places, séjour de longue durée']+df['Nombre de places, séjour de courte durée']
# estimation 15m2 par lits
df['Esti_m2']=df.Lits*15
```

```python
df['C_couvert_LAMAL']=df['Coût des soins couverts LAMal par journée de soins (en francs)']+df['Coût des autres prestations LAMal par journée de soins (en francs)']
df['pourcentage_C_couvert_LAMAL']=df['Coût des soins couvertes LAMal (%)']+df['Autres prestations couvertes LAMal (%)']
# Pourcentage MAsse Salariale sur Coûts totaux
df['MSsurC'] = df['Charges salariales  (en milliers de francs)']/df['Ensemble des coûts (en milliers de francs)']*100
```

```python
df.describe()
```

# 1. Sélection des EMS étant SA ou Fondation

```python
SA = df[df['Forme juridique'] == 'Société anonyme']
Fdt = df[df['Forme juridique'] == 'Fondation (CS art. 80 & ss)']
SAetFound = pd.concat([SA, Fdt])
SAetFound.describe()
```

```python
display_interative_plot(df, "Lits", "Benefice")
```

## 1.1 Les 10 EMS faisant le plus gros Profit

```python
# 10 EMS ayant le plus gros CA
SAetFound["Benefice"] = SAetFound["Benefice"].astype(float)
SAetFound.nlargest(10,"Benefice")
#SAetFound.info()
```

```python
french_cantons = ["JU", "FR", "VA", "GE", "NE", "VD"]
Romandie = SAetFound[SAetFound['Canton'].isin(french_cantons)]
```
